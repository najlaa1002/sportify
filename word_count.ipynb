The rdd.flatMap is a command that creates a new rdd by performing a function to the initial rdd. The results of this function are stored in lists and then flattened into a single one with the

```python
# Initialize a word counter by creating a tuple with word and cound of 1
words = rdd.flatMap(lambda lines: lines.split(' ')) \
                    .map(lambda word: (word, 1))

for w in words.collect():
    print(w)
```

```python
# a. count the occurence of each word


word_counts = rdd.flatMap(lambda lines: lines.split(' ')) \
                 .map(lambda word: (word, 1)) \
                 .reduceByKey(lambda x, y: x + y)

print("Sample word counts:")
for word, count in word_counts.take(10):
    print(f"'{word}': {count}")
```

```python
# b. a common first step in text analysis, change all capital letters to lower case


lowercase_words = rdd.flatMap(lambda lines: lines.split(' ')) \
                     .map(lambda word: word.lower()) \
                     .map(lambda word: (word, 1)) \
                     .reduceByKey(lambda x, y: x + y)

print("Sample lowercase word counts:")
for word, count in lowercase_words.take(10):
    print(f"'{word}': {count}")
```

```python
# c. eliminate the stop words.
stop_words = {
    'a', 'an', 'and', 'are', 'as', 'at', 'be', 'by', 'for', 'from',
    'has', 'he', 'in', 'is', 'it', 'its', 'of', 'on', 'that', 'the',
    'to', 'was', 'will', 'with', 'had', 'have', 'his', 'her', 'him',
    'she', 'they', 'we', 'you', 'i', 'me', 'my', 'your', 'their',
    'this', 'these', 'those', 'but', 'or', 'not', 'no', 'so', 'up',
    'out', 'if', 'about', 'who', 'what', 'where', 'when', 'why', 'how'
}

words_no_stopwords = rdd.flatMap(lambda lines: lines.split(' ')) \
                        .map(lambda word: word.lower()) \
                        .filter(lambda word: word not in stop_words and word != '') \
                        .map(lambda word: (word, 1)) \
                        .reduceByKey(lambda x, y: x + y)

print("Sample words without stop words:")
for word, count in words_no_stopwords.take(10):
    print(f"'{word}': {count}")
```

```python
# d. sort in alphabetical order
alphabetical_words = words_no_stopwords.sortByKey()
for word, count in alphabetical_words.take(10):
    print(f"'{word}': {count}")
```

```python
# e. sort descending by word frequency

frequency_sorted = words_no_stopwords.map(lambda x: (x[1], x[0])) \
                                     .sortByKey(ascending=False) \
                                     .map(lambda x: (x[1], x[0]))

print("Top 15 most frequent words:")
for word, count in frequency_sorted.take(15):
    print(f"'{word}': {count}")
```

```python
# f. remove punctuations and blank spaces
import string

def clean_word(word):
    # Remove punctuation and convert to lowercase
    cleaned = word.lower().strip()
    # Remove punctuation characters
    cleaned = ''.join(char for char in cleaned if char not in string.punctuation)
    return cleaned

clean_words = rdd.flatMap(lambda lines: lines.split(' ')) \
                 .map(clean_word) \
                 .filter(lambda word: word != '' and word not in stop_words) \
                 .map(lambda word: (word, 1)) \
                 .reduceByKey(lambda x, y: x + y) \
                 .map(lambda x: (x[1], x[0])) \
                 .sortByKey(ascending=False) \
                 .map(lambda x: (x[1], x[0]))

print("Top 20 words after full cleaning:")
for word, count in clean_words.take(20):
    print(f"'{word}': {count}")
```
